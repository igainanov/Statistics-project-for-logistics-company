Задание 1.

Менеджер логистической компании решил провести АБ-тест, выбрав в качестве
метрики время доставки курьерами посылок. Для контрольной группы курьеров
оставили предыдущий интерфейс навигации в приложении, а для тестовой
группы внедрили новую опцию, позволяющую комбинировать несколько видов
транспорта.
Полученные результаты времени доставки в минутах для тестовой и
контрольной групп представлены в csv-файлах times_couriers_control.csv и
times_couriers_test.csv.
Проверьте гипотезу менеджера о том, что внедрённая опция приложения
уменьшила время доставки.

Нулевая гипотеза (H0) - внедрение новой опции не влияет на время доставки, а альтернативная гипотеза (H1) - внедрение новой опции сокращает время доставки. Уровень статистической значимости (alpha) берем 0.05.

```python
#Загрузим данные из файлов:

from scipy import stats as st

import pandas as pd

# Загрузка данных
control_df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/DA/times_couriers_control.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/DA/times_couriers_test.csv')
```

```python
print(control_df.columns)
```

```python
print(test_df.columns)
```

```python
# Вычисляем средние продолжительности доставки для контрольной и тестовой групп
mean_control = control_df['time'].mean()
mean_test = test_df['time'].mean()

# Проведем t-тест для независимых выборок
results = st.ttest_ind(test_df['time'], control_df['time'], alternative = "less")

# Выведем результаты
print(f"Среднее время доставки в контрольной группе: {mean_control:.2f}")
print(f"Среднее время доставки в тестовой группе: {mean_test:.2f}")
print(f"t-статистика: {results.statistic:.2f}")
print(f"p-value: {results.pvalue:.5f}")

# Определим, была ли разница статистически значимой
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Не получилось отвергнуть нулевую гипотезу, т.к. среднее время доставки в тестовой группе не уменьшилось, а наоборот увеличилось. Сформулируем гипотезы: нулевая гипотеза (H0) - внедрение новой опции не влияет на время доставки, а альтернативная гипотеза (H1) - внедрение новой опции увеличивает время доставки. Уровень статистической значимости (alpha) берем 0.05.

```python
# Проведем t-тест для независимых выборок
results = st.ttest_ind(test_df['time'], control_df['time'], alternative = "greater")

# Выведем результаты
print(f"t-статистика: {results.statistic:.2f}")
print(f"p-value: {results.pvalue:.5f}")

# Определим, была ли разница статистически значимой
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Задание 2.

В этой же компании с 1 мая часто заказывающим клиентам предоставили скидку, действующую на все заказы до конца календарного месяца, если в этом месяце клиент уже сделал четыре заказа (то есть начиная с пятого заказа). В файле clients_orders.csv представлена информация о количестве заказов, которые одни и те же клиенты сделали в апреле и мае. Проверьте гипотезу о том, что клиенты стали делать больше заказов после введения скидки. Подсказка: это зависимая (парная) выборка.


Нулевая гипотеза (H0) - внедрение акционной скидки в мае не привело к увеличению количества заказов, а альтернативная гипотеза (H1) - внедрение акционной скидки в мае привело к увеличению количества заказов. Уровень статистической значимости (alpha) берем 0.05.

```python
#загрузка csv файла
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/DA/clients_orders.csv')
```

```python
print(data.columns)
```

```python
import pandas as pd
from scipy.stats import ttest_rel

# Проведение t-теста для зависимых парных выборок
t_stat, p_value = ttest_rel(data['orders_may'], data['orders_april'],
                       alternative = "greater")

# Вывод результата
print(f"t-статистика: {t_stat:.2f}")
print(f"p-value: {p_value:.5f}")

# Оцениваем статистическую значимость
alpha = 0.05  # Уровень статистической значимости

# Интерпретация результатов
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Задание 3.

В файле couriers_education.csv представлена анонимизированная информация о курьерах компании: образование ( education ) и количество лет работы в компании ( experience ).

Переменная education принимает значения:

NS — no school — нет аттестата об окончании школы

MS — middle school — закончены 9 классов школы

S — school — среднее образование: закончена средняя школа

SSE — secondary special education — среднее специальное образование (ПТУ / колледж)

HE — higher education — высшее образование.

Переменная experience принимает значения:

0-1 — до одного года работы в компании

1-3 — от одного до трёх лет работы

3-5 — от трёх до пяти лет работы

5- — от пяти лет работы.

Проверьте гипотезу о том, что стаж работы в компании зависит от уровня образования.

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Загружаем данные из CSV файла
data = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/DA/couriers_education.csv')

# Создаем кросс-таблицу для уровня образования и стажа работы
cross_table = pd.crosstab(data['education'], data['experience'])

# Выполняем тест хи-квадрат
chi2, p_value, _, _ = chi2_contingency(cross_table)

# Выводим результаты теста
print(f"Хи-квадрат статистика: {chi2:.2f}")
print(f"p-value: {p_value:.5f}")

# Оцениваем статистическую значимость
alpha = 0.05  # Уровень статистической значимости
if p_value < alpha:
    print("Отвергаем гипотезу о независимости уровня образования и стажа работы.")
else:
    print("Не отвергаем гипотезу о независимости уровня образования и стажа работы.")
```

Задание 4.

Постройте 95%-е доверительные интервалы по выборкам из задания 1 для
среднего времени доставки курьерами, использующими и не использующими
новую опцию комбинирования нескольких видов транспорта в приложении.

```python
import pandas as pd
from scipy import stats

# Загрузка данных
control_df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/DA/times_couriers_control.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/DA/times_couriers_test.csv')

# Уровень доверия
confidence_level = 0.95

# Доверительный интервал для контрольной группы
control_mean = control_df['time'].mean()
control_sem = stats.sem(control_df['time'])
control_confidence_interval = stats.t.interval(confidence_level, len(control_df)-1, loc=control_mean, scale=control_sem)

# Доверительный интервал для тестовой группы
test_mean = test_df['time'].mean()
test_sem = stats.sem(test_df['time'])
test_confidence_interval = stats.t.interval(confidence_level, len(test_df)-1, loc=test_mean, scale=test_sem)

# Вывод результатов
print(f"Доверительный интервал для контрольной группы: {control_confidence_interval}")
print(f"Доверительный интервал для тестовой группы: {test_confidence_interval}")
```
